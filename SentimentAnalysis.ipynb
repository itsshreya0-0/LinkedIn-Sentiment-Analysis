LINKED LIST

```python
# LINKED LIST
class Node:
    def __init__(self, data):
        self.data = data #Stores info about one post
        self.next = None #Pointer to the next node

class LinkedList:
    def __init__(self):
        self.head = None #head initially empty

    def insert(self, data):
        newnode = Node(data)
        if self.head is None:
            self.head = newnode
        else:
            temp = self.head
            while temp.next is not None:
                temp = temp.next
            temp.next = newnode

    def display(self):
        temp = self.head
        while temp is not None:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")
```

SORTING NAMES BY USING MERGE SORT

```python

# SORTING NAMES BY MERGE SORT
def merge(left, right):
    merged = []
    i = j = 0

    # Comparing elements one by one
    while i < len(left) and j < len(right):
        if left[i]["name"].lower() < right[j]["name"].lower():
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1

    while i < len(left):
        merged.append(left[i])
        i += 1
    while j < len(right):
        merged.append(right[j])
        j += 1
    return merged

# Merge Sort function
def merge_sort(data):
    if len(data) <= 1:
        return data

     # Find the middle index and split
    mid = len(data) // 2
    left_half = merge_sort(data[:mid])
    right_half = merge_sort(data[mid:])
    return merge(left_half, right_half)
```

BINARY SEARCH OF NAMES

```python
# --- Binary Search Function ---
def binary_search(profiles, target):
    low = 0
    high = len(profiles) - 1

    while low <= high:
        mid = (low + high) // 2
        name = profiles[mid]["name"].lower()

        if name == target.lower():
            return profiles[mid]
        elif name < target.lower():
            low = mid + 1
        else:
            high = mid - 1

    return None
```

```python
# CONVERT ACTIVITIES TO LINKED LIST
def convert_post_activities(profiles):
    for profile in profiles:
        if isinstance(profile.get("post_activities"), list):
            new_linked_list = LinkedList()
            for activity in profile["post_activities"]:
                new_linked_list.insert(activity) # Corrected method name
            profile["post_activities"] = new_linked_list
    return profiles


# === MAIN PROGRAM ===
profiles = merge_sort(profiles)
profiles = convert_post_activities(profiles)

print("Profiles sorted by name:")
for p in profiles:
    print(p["name"], ":", p["title"])
```

```python
import numpy as np
from scipy.sparse import csr_matrix

def build_sparse_skill_matrix(profiles):
    skills = {}
    data, rows, cols = [], [], []

    for i, p in enumerate(profiles):
        for skill, val in p["skills_and_endorsements"].items():
            if skill not in skills:
                skills[skill] = len(skills)
            rows.append(i)
            cols.append(skills[skill])
            data.append(val)

    sparse_matrix = csr_matrix((data, (rows, cols)), shape=(len(profiles), len(skills)))
    return sparse_matrix, skills

skill_sparse, skill_index = build_sparse_skill_matrix(profiles)

print("Sparse shape:", skill_sparse.shape)
print("Non-zero endorsements:", skill_sparse.nnz)

```

```python
import networkx as nx
import matplotlib.pyplot as plt

def directed_follow_graph(profiles):
    # Create a directed graph
    G = nx.DiGraph()

    # Add all people as nodes
    for p in profiles:
        G.add_node(p["name"])

    # Add edges (arrows) for "who follows whom"
    for p in profiles:
        for person in p["follows"]:
            G.add_edge(p["name"], person)

    # Automatically arrange the nodes
    pos = nx.spring_layout(G)

    # Draw the graph
    plt.figure(figsize=(12, 8))
    nx.draw(
        G, pos,
        with_labels=True,
        node_color="lightblue",
        node_size=800,
        arrows=True,
        arrowsize=15,
        font_size=10,
        font_weight="bold"
    )

    plt.title("Who Follows Whom (Directed Graph)")
    plt.axis("off")
    plt.show()

directed_follow_graph(profiles)
```

STACK IMPLEMENTATION IN RECENT PROFILE VIEWERS

```python
from datetime import datetime
from typing import List

class Viewer:
    def __init__(self, name: str, timestamp: str):
        self.name = name
        self.timestamp = datetime.strptime(timestamp, "%Y-%m-%d %H:%M:%S")

    def __repr__(self):
        return f"{self.name} at {self.timestamp.strftime('%Y-%m-%d %H:%M:%S')}"

class ViewerStack:
    def __init__(self):
        self.stack: List[Viewer] = []

    def push(self, viewer: Viewer):
        self.stack.append(viewer)

    def pop(self):
        if self.stack:
            return self.stack.pop()
        return None

    def peek(self):
        if self.stack:
            return self.stack[-1]
        return None

    def display_recent_viewers(self):
        print("Recent Profile Viewers (Most Recent on Top):")
        for viewer in reversed(self.stack):
            print(viewer)

class Profile:
    def __init__(self, name):
        self.name = name
        self.viewers = ViewerStack()

    def add_viewer(self, name, timestamp):
        self.viewers.push(Viewer(name, timestamp))

    def show_recent_viewers(self):
        print(f"\nProfile: {self.name}")
        self.viewers.display_recent_viewers()


# Example usage
profile = Profile("Rhithika Rajkumar")

profile.add_viewer("Arthy", "2025-10-11 14:23:45")
profile.add_viewer("Prithika", "2025-10-11 13:50:10")
profile.add_viewer("Selva Prabhu", "2025-10-11 12:35:22")
profile.add_viewer("Karan Agarwal", "2025-10-11 11:12:05")

profile.show_recent_viewers()
```

```python
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
def get_profile_text(profile):
    text_parts = [
        profile.get("about", ""),
        profile.get("headline", ""),
        profile.get("experience", ""),
    ]
    # Handle post_activities which is now a LinkedList
    post_activities_text = ""
    if isinstance(profile.get("post_activities"), LinkedList):
        temp = profile["post_activities"].head
        activities = []
        while temp is not None:
            activities.append(temp.data)
            temp = temp.next
        post_activities_text = " ".join(activities)
    elif isinstance(profile.get("post_activities"), list): # Keep original list handling just in case
         post_activities_text = " ".join(profile.get("post_activities", []))

    text_parts.append(post_activities_text)
    return " ".join(text_parts)

def analyze_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity

def sentiment_label(score):
    if score > 0.2:
        return "Positive"
    elif score < -0.2:
        return "Negative"
    else:
        return "Neutral"
```

```python
results = []

for profile in profiles:
    text = get_profile_text(profile)
    score = analyze_sentiment(text)
    label = sentiment_label(score)
    results.append({
        "name": profile["name"],
        "sentiment_score": round(score, 2),
        "sentiment_label": label
    })

for r in results:
    print(f"{r['name']}: {r['sentiment_label']} ({r['sentiment_score']})")

```

```python
def generate_suggestions(text, score):
    suggestions = []

    confidence_keywords = [
        "led", "developed", "built", "improved", "optimized",
        "achieved", "designed", "managed", "launched", "increased"
    ]

    weak_tone_keywords = [
        "trying", "learning", "exploring", "interested",
        "hoping", "basic", "beginner"
    ]

    job_search_keywords = [
        "open to work", "exploring opportunities", "seeking",
        "looking for roles", "career switch"
    ]

    technical_keywords = [
        "ai", "machine learning", "cloud", "data science", "ux",
        "react", "aws", "python", "analytics", "devops"
    ]

    leadership_keywords = [
        "leader", "manager", "founder", "architect", "head",
        "leadership", "strategy"
    ]

    text_lower = text.lower()

    if score < 0:
        suggestions.append("Your profile tone appears slightly weak or uncertain. Strengthen it with confident, impact-driven language.")

        for word in weak_tone_keywords:
            if word in text_lower:
                suggestions.append(f"Replace soft wording like '{word}' with strong action verbs such as 'led', 'built', or 'delivered'.")
                break

        if any(word in text_lower for word in job_search_keywords):
            suggestions.append("Since you are open to opportunities, clearly highlight your key achievements and measurable results.")

        if not any(word in text_lower for word in technical_keywords):
            suggestions.append("Consider adding more technical or domain-specific skills to strengthen your professional identity.")

        suggestions.append("Add 1–2 quantified achievements (e.g., 'increased performance by 30%').")

    else:
        suggestions.append("Your profile sentiment is strong and positive. Keep showcasing your achievements and career growth.")

        if any(word in text_lower for word in leadership_keywords):
            suggestions.append("Your leadership presence is strong. Consider sharing more strategic insights or industry thought leadership posts.")

        if any(word in text_lower for word in technical_keywords):
            suggestions.append("Your technical positioning is excellent. Continue sharing real-world use cases and problem-solving examples.")

        if any(word in text_lower for word in job_search_keywords):
            suggestions.append("Since you're open to opportunities, make sure your most impactful achievements appear at the top of your profile.")

        suggestions.append("Consistent posting about your work will further enhance your professional visibility.")

    return suggestions

```

```python
def get_profile_text(profile):
    text_parts = []

    def safe_str(value):
        if value is None:
            return ""
        return str(value)


    text_parts.append(safe_str(profile.get("about")))
    text_parts.append(safe_str(profile.get("headline")))
    text_parts.append(safe_str(profile.get("experience")))


    post_activities = profile.get("post_activities")

    posts = []


    if isinstance(post_activities, (list, tuple)):
        for post in post_activities:
            posts.append(safe_str(post))


    elif hasattr(post_activities, "head"):
        current = post_activities.head
        while current:
            posts.append(safe_str(current.data))
            current = current.next


    elif isinstance(post_activities, str):
        posts.append(post_activities)


    else:
        posts.append("")

    text_parts.append(" ".join(posts))


    final_text = " ".join(text_parts).strip()

    if not final_text:
        final_text = "neutral professional profile"

    return final_text
```

```python
#FORCEFULL IMPLEMETATION OF GRAPHS
import matplotlib.pyplot as plt

class ScoreGraph:
    def __init__(self):
        # adjacency list: node -> list of neighbors
        self.adj = {}
        # store score for each node
        self.scores = {}

    def add_node(self, node_name, score):
        self.scores[node_name] = score
        if node_name not in self.adj:
            self.adj[node_name] = []

    def add_edge(self, from_node, to_node):
        if from_node not in self.adj:
            self.adj[from_node] = []
        self.adj[from_node].append(to_node)

    def get_path_scores(self, start_node):
        """Simple DFS from start_node to collect nodes in order, with scores."""
        order = []
        visited = set()

        def dfs(u):
            visited.add(u)
            order.append(u)
            for v in self.adj.get(u, []):
                if v not in visited:
                    dfs(v)

        dfs(start_node)
        score_list = [self.scores[n] for n in order]
        return order, score_list
```

```python
# --- RUN SENTIMENT ANALYSIS TWICE ---

print("Running initial sentiment analysis...")
s1 = get_individual_sentiment(profiles)   # Initial score

print("\nRunning sentiment analysis after improvements...")
s2 = get_individual_sentiment(profiles)   # Improved score

# --- BUILD GRAPH DATA STRUCTURE ---

g = ScoreGraph()

# Add nodes for the two states
g.add_node("S1", s1)   # Initial score
g.add_node("S2", s2)   # Improved score

# Add directed edge from S1 -> S2 (progression)
g.add_edge("S1", "S2")

# Get traversal order and corresponding scores from the graph
order, score_list = g.get_path_scores("S1")

# --- VISUALIZE USING THE GRAPH DATA ---

plt.figure(figsize=(7, 5))
plt.plot(order, score_list, marker='o', linewidth=2)
plt.title("Sentiment Score Progression (Graph-based)")
plt.xlabel("States (Graph Nodes)")
plt.ylabel("Sentiment Score")
plt.grid(True)

# Show score values above each node
for i, score in enumerate(score_list):
    plt.text(i, score + 0.01, f"{score:.2f}", ha='center')

plt.show()
```
